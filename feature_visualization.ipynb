```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


train = pd.read_csv("train_sample2.csv",header=None)
test = pd.read_csv("test_sample2.csv",header=None)
data_311=pd.concat([train[train[19]==311],test[test[19]==311]],sort=False)
data_348=pd.concat([train[train[19]==348],test[test[19]==348]],sort=False)
def plotdata(dataframe,title):# groupby weekend
    groups = dataframe.groupby(23)
    fig, ax = plt.subplots()
    ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
    for name, group in groups:
        ax.plot(group[22], group[35], marker='o', linestyle='', ms=7, label=name,alpha=0.3)
    ax.legend()
    ax.set_xlabel("hour")
    ax.set_ylabel("p20 total waiting time")
    ax.set_title(title)
    plt.show()
plotdata(data_348,"Stratified by weekend (1) or not (0)")
plotdata(data_311,"Stratified by weekend (1) or not (0)")

def plotdatabymonth(dataframe,title):
    groups = dataframe.groupby(24)
    fig, ax = plt.subplots()
    ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
    for name, group in groups:
        ax.plot(group[22], group[35], marker='o', linestyle='', ms=7, label=name,alpha=0.3)
    ax.legend()
    ax.set_xlabel("hour")
    ax.set_ylabel("p20 total waiting time")
    ax.set_title(title)
    plt.show()
    
plotdatabymonth(data_348[data_348[23]==0],"p20 total waiting time stratified by month")

plotdatabymonth(data_311[data_311[23]==0],"p20 total waiting time stratified by month")
```
